Reduction operation

```python
j = torch.randint(size=(2,3),low=1, high=10, dtype=torch.float32)
j
```

```python
# sum
torch.sum(j)
# sum along columns
torch.sum(j, dim=0)
# sum along rows
torch.sum(j, dim=1)
```

```python
#mean
torch.mean(j)
# mean along columns
torch.mean(j, dim=0)
# mean along rows
torch.mean(j, dim=1)
```

```python
#median, min, max, product, Sd,vairance
torch.median(j)
torch.min(j)
torch.max(j)
torch.prod(j)
torch.std(j)
torch.var(j)
```

```python
#argmax----> finding index number of largest value
torch.argmax(j)
#argmin--->finding minimum number index
torch.argmin(j)
```
